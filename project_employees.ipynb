```python
#Categorical vs continuous (numerical) columns : Boxplot, Barplot
#Plots used are: bar plot and box plot
sns.barplot(x='WorkLifeBalance',y='TotalWorkingYears',data=hr_data) #barplot between WorkLifeBalance and TotalWorkingYears
sns.boxplot(x='Education',y='Age',data=hr_data, palette='rainbow') #boxplot between Education vs Age
```

```python
#Continuous vs continuous columns: scatter plot
hr_data.plot.scatter(x="DailyRate",y="Age")
plt.show()
```

```python
#Categorical vs Categorical columns: Group By (Sum, Count ,Value Count)
pd.crosstab(hr_data.Education, hr_data.DistanceFromHome, margins=True,values=hr_data.TotalWorkingYears, aggfunc=pd.Series.count)
# need to summarize on x and y categorical variables
# 3 any other aggregation function can be used based on column type
```

```python
#Categorical vs Categorical columns: #2.Crosstab Plot(pie chart)   
# Crosstabs
cross_tab_prop = pd.crosstab(index = hr_data['DistanceFromHome'], columns = hr_data['Education'], normalize = "index")
cross_tab_prop
```

# Discrete Distribution

```python
#A discrete distribution is a probability distribution that depicts the occurrence of discrete 
#(individually countable) outcomes, such as 1, 2, 3... or zero vs. one.
#For example, the possible values for the random variable X that represents the number of heads that can occur when a coin is 
#tossed twice are the set {0, 1, 2} and not any value from 0 to 2 like 0.1 or 1.6.
#Requirement The probabilities of random variables must have discrete (as opposed to continuous) values as outcomes.
```

# Continuous probability distributions

```python
#these distributions model the probabilities of random variables that can have any possible outcome. For example, the possible 
#values for the random variable X that represents weights of citizens in a town which can have any value like 34.5, 47.7, etc.,

#Examples: Normal, Student’s T, Chi-square, Exponential, etc.,
# a continuous probability distribution can contain outcomes that have any value, including indeterminant fractions.
```

# Central Limit Theorem

```python
#The Central Limit Theorem(CLT) states that for any data, provided a high number of samples have been taken.
#1 Sampling Distribution Mean(uₓ-) = Population Mean(u)
#2 Sampling distribution’s standard deviation (Standard error) = sigma/sq.root(n) ~S/sq.root(n)
#3 For n > 30, the sampling distribution becomes a normal distribution.
```

```python
# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
```

```python
df = pd.read_csv('HR-Employee-Attrition.csv')

#Printing the dataset
df
```

```python
# Extracting Age Column from the dataset
df = df[['Age']]
df
```

```python
# Plotting the distribution graph using Seaborn Library
sns.distplot(df.Age) 
plt.show()
#A distplot plots a univariate distribution of observations. The distplot() function combines the matplotlib hist function with 
#the seaborn kdeplot() and rugplot() functions.
```

```python
# Mean of the Age Column
df.Age.mean()
```

```python
# Standard eviation of the Age Column
df.Age.std()
```

```python
# Taking sample size of 30
samp_size = 30
# Taking the sample and calculating it's mean
df.Age.sample(samp_size).mean()
```

```python
# Creating multiple number of samples and calculating it's means
sample_means = [df.Age.sample(samp_size).mean() for i in range(1000)]
# Storing the mean values in a Series using Pandas
sample_means = pd.Series(sample_means)
# Verifying the total number of samples collected
len(sample_means)
```

```python
# Plotting the density for the sample means.
sns.distplot(sample_means)
plt.show()
```

```python
# Mean value of all the 1000 Samples
sample_means.mean()
```

```python
# Standard Deviation of all the 1000 Samples
sample_means.std()
```

```python
df.Age.std()/np.sqrt(samp_size)
```
